```python
# Lowercasing and uppercasing a data

# Import pandas package
import pandas as pd

# Define a dictionary containing employee data
data = {
    "Name": ["Jai", "Princi", "Gaurav", "Anuj"],
    "Age": [27, 24, 22, 32],
    "Address": ["Delphi", "Kanpur", "Allahabad", "Kannauj"],
    "Qualification": ["Msc", "MA", "MCA", "Phd"]
}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)

# Converting and overwritting values in column
df["Name"] = df["Name"].str.lower()
df
```

```python
# Importing pandas package
import pandas as pd

# Making data frame from csv file
data = pd.read_csv("datasets/nba.csv")

# Converting and overwritting values in column
data["Team"] = data["Team"].str.upper()

# Display
data
```

```python
# Splitting and replacing a data

# Importing pandas module
import pandas as pd

# Define a dictionary containing employee data
data = {
    "Name": ["Jai", "Princi", "Gaurav", "Anuj"],
    "Age": [27, 24, 22, 32],
    "Address": ["Delphi", "Kanpur", "Allahabad", "Kannauj"],
    "Qualification": ["Msc", "MA", "MCA", "Phd"]
}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)

# Dropping null value columns to avoid errors
df.dropna(inplace=True)

# New data frame with split value columns
df["Address"] = df["Address"].str.split("a", n=1, expand=True)

# Display
df
```

```python
# Importing pandas module
import pandas as pd

# Reading csv file from url
data = pd.read_csv("datasets/nba.csv")

# Overwritting column with replaced value of age
data["Age"] = data["Age"].replace(25.0, "Twenty five")

# Creating a filter for age column where age = "Twenty five"
filtered_ages = data["Age"] == "Twenty five"

# Printing only filtered columns
data.where(filtered_ages).dropna()[:5]
```

```python
# Concatenation of Data

# Importing pandas module
import pandas as pd

# Define a dictionary containig employee data
data = {
    "Name": ["Jai", "Princi", "Gaurav", "Anuj"],
    "Age": [27, 24, 22, 32],
    "Address": ["Delphi", "Kanpur", "Allahabad", "Kannauj"],
    "Qualification": ["Msc", "MA", "MCA", "Phd"]
}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)

# Making copy of address column
new = df["Address"].copy()

# Concatenating address with name column overwriting name column
df["Name"] = df["Name"].str.cat(new, sep=", ")

# Display
df
```

```python
# Removing whitespaces of data

# Importing pandas module
import pandas as pd

# Define a dictionary containing employee data
data = {
    "Name": ["Jai", "Princi", "Gaurav", "Anuj"],
    "Age": [27, 24, 22, 32],
    "Address": ["Delphi ", "  Kanpur", " Allahabad", "Kannauj    "],
    "Qualification": ["Msc", "MA", "MCA", "Phd"]
}

# Convert the dictionary into DataFrame
df = pd.DataFrame(data)

# Replacing whitespaces
df["Address"] = df["Address"].str.strip()
df
```
